```python
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]

def my_median(ages):
    ages.sort() #Sort the list to find the median value
    print('Sorted list:', ages)

    if len(ages)%2 == 0 : #Find if the length of the list is even or odd
        median_val_1 = int(len(ages)/2 - 1)
        median_val_2 = int(len(ages)/2)
        median_even_value = (median_val_1+median_val_2)/2 #Forumla for even vals
        print('Even value is : ',ages[median_even_value])
    else:
        median_odd_value = int((len(ages)+1)/2 -1 ) #Odd value forumla
        print('Odd value is : ',ages[median_odd_value])

my_median(ages)

list_count = len(ages)-1 
greater_average = round((list_count)*0.9) #To find the index value
print('Position of the greater value is: ',greater_average)
print('The age greater than 90% of the other ages is: ',ages[greater_average]) #To print the element in the position
```

```python
num = 5

#same logic as Pascal's triangle

i=1 
while num>0: # to check if the number is positive
    while i<=num: # outer loop checks for the row
        j=1
        while (j<=i): #inner loop checks for the columns
            print("*", end = " ")
            j+=1
        print("") # to print the next line
        i+=1
    break #to break out of the loop

k=1
while k<=num-1: #reverse printing of the same logic as above
    l = num
    while l-1>=k: # to print one element less than the middle element . i.e, 5-1 = 4(****)
        print("*", end = " ")
        l-=1
    print("")
    k+=1
```

```python
def find_all(string, sstring):
    string_list = [] # Empty list 
    string_length = len(string)
    string_index = 0 # Assigning the Index value of the string

    while string_index < string_length: # Loop will continute until the index value <= to length of the string
        i = string.find(sstring,string_index) # Find function returns the first instance of the substring
        if i == -1: # if no occurence found, return empty list
            return string_list
        string_list.append(i) # Attach the index value to the new list
        string_index = i+1 #Increment the value of the index value
    return string_list

print(find_all('helloworldhelloPythonhelloc++hellojava','hello'))
```

```python
def sum_digits(n):
    r=0
    while n:
        r, n = r + n % 10, n // 10
    return r

def check_studentid(studentid): 
    x = sum_digits(studentid)
    if x % 2 == 0:
        print ('version I')
    else :
        print('version II ')

check_studentid (220156351)
```
